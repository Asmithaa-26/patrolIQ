```python
import pandas as pd

# Load raw data
df = pd.read_csv(r"C:\Users\asmis\OneDrive\Desktop\patroliq_V1\patroliq\data\raw\Crimes_-_2001_to_Present_20260205.csv")

print("Original shape:", df.shape)

# Datetime conversion
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Keep latest 500K records
df = df.sort_values("Date", ascending=False).head(500_000)

# Drop missing geo coordinates
df = df.dropna(subset=["Latitude", "Longitude"])

print("Cleaned shape:", df.shape)

# Save cleaned data
df.to_csv(r"C:\Users\asmis\OneDrive\Desktop\patroliq_V1\patroliq\data\raw\Crimes_-_2001_to_Present_20260205.csv", index=False)

print("Saved to data/processed/cleaned_data.csv")
```
