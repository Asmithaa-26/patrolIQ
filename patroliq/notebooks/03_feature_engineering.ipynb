```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv(r"C:\Users\asmis\OneDrive\Desktop\patroliq_V1\patroliq\data\processed\cleaned_data.csv")
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Hour"] = df["Date"].dt.hour
df["Day_of_Week"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month
df["Is_Weekend"] = df["Day_of_Week"].isin(["Saturday", "Sunday"]).astype(int)
```

```python
def season(month):
    if month in [12,1,2]: return "Winter"
    if month in [3,4,5]: return "Spring"
    if month in [6,7,8]: return "Summer"
    return "Fall"

df["Season"] = df["Month"].apply(season)
```

```python
severity_map = {
    "HOMICIDE": 5,
    "CRIM SEXUAL ASSAULT": 5,
    "KIDNAPPING": 5,
    "ROBBERY": 4,
    "ASSAULT": 3,
    "BATTERY": 3,
    "BURGLARY": 3,
    "THEFT": 2,
    "CRIMINAL DAMAGE": 2
}

df["Crime_Severity_Score"] = df["Primary Type"].map(severity_map).fillna(1)
```

```python
le = LabelEncoder()

df["Primary_Type_Encoded"] = le.fit_transform(df["Primary Type"])
df["Location_Encoded"] = le.fit_transform(df["Location Description"])
df["Season_Encoded"] = le.fit_transform(df["Season"])
```

```python
model_features = [
    "Latitude", "Longitude",
    "Hour", "Month",
    "Crime_Severity_Score",
    "Primary_Type_Encoded",
    "Location_Encoded",
    "Is_Weekend"
]

X = df[model_features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```
